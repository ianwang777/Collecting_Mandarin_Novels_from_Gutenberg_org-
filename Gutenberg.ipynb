```python
from selenium.webdriver.chrome.service import Service
from selenium import webdriver
from selenium.common.exceptions import TimeoutException
from selenium.common.exceptions import NoSuchElementException
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.common.keys import Keys
from time import sleep
import operator
import json
import os
import re

reg_ch = r'^[\u4E00-\u9FFF\u3002\uff1f\uff01\uff0c\u3001\uff1b\uff1a\u201c\u201d\u2018\u2019\uff08\uff09\u300a\u300b\u3008\u3009\u3010\u3011\u300e\u300f\u300c\u300d\u3014\u3015\u2026\u2014\uff5e]'
reg_ch_not = r'[^\u4E00-\u9FFF\u3002\uff1f\uff01\uff0c\u3001\uff1b\uff1a\u201c\u201d\u2018\u2019\uff08\uff09\u300a\u300b\u3008\u3009\u3010\u3011\u300e\u300f\u300c\u300d\u3014\u3015\u2026\u2014\uff5e\s]'
reg_en = r'[a-zA-Z]'

options = webdriver.ChromeOptions()
options.add_argument("--start-maximized")         #最大化視窗
options.add_argument("--disable-popup-blocking") #禁用彈出攔截

#selenium 3版
# path = "../chromedriver"
# driver = webdriver.Chrome(path,options=options)

# selenium 4版
service = Service(executable_path="../../chromedriver.exe")
driver = webdriver.Chrome(
    options = options,
    service = service
)
driver.get('https://www.gutenberg.org/browse/languages/zh')
sleep(1)

# 創個資料夾 Gutenberg 裝下載好的文字檔
if not os.path.isdir('Gutenberg'):
    os.mkdir('Gutenberg')
```

```python
books = driver.find_elements(By.CSS_SELECTOR, 'li.pgdbetext a')
i = 1
name_and_link = []
for e in books:
#     print(i)
        
    book_name = e.text
#     print(book_name)

    book_no = e.get_attribute('href').split('/')[-1]
    prefix = 'https://www.gutenberg.org/files/'
    link = prefix+book_no+'/'+book_no+'-0.txt'
#     print(link)

    name_and_link.append({
        "name":book_name,
        "link":link
    })
           
#     print("="*20)
    i = i+1

print(len(name_and_link))
```

```python
i = 1
for e in name_and_link:
    print(i)
    print(e['link'])
    driver.get(e['link'])
    
    # 如果找不到 pre 標籤表示 error 404, 跳過這次迴圈
    try:
        content = driver.find_element(By.CSS_SELECTOR, 'pre')
    except NoSuchElementException:
       continue
    
    
    #寫檔案, 用regex判斷書名內有英文字的就不抓
    match_name = re.search(reg_en, e['name'])
    if match_name == None:
    
        book_file = "Gutenberg/"+e['name'].replace("\n"," ")+".txt" #e['name'].replace("\n"," ") 為了處理書名有換行的狀況
        
        with open(book_file, "w", encoding='utf-8') as file:
            match_content1 = re.sub(reg_ch_not, '', content.text) #去掉非中文
            # match_content2 = re.sub(r"[ ]+(?=\n)", '', match_content1) #去掉文章前後的空白
            match_content3 = re.sub(r"(?<!\S)\n(?=\s)", '', match_content1) #去掉文章前後的換行
            
            start = match_content3.find('\n')+1
            end = match_content3.rfind('\n')
            file.write(match_content3[ start : end])
    
    print("="*20)
    i=i+1
```
